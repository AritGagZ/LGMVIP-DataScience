***Task 2 - Image to Pencil Sketch with Python***

```python
import cv2
import matplotlib.pyplot as plt
```

```python
#Read the image
img = cv2.imread("E:/Arit/Images/My Photography/my photography #Redminote3/bug.jpg")
RGB_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(RGB_img)
plt.axis(False)
plt.show()
```

```python
#Convert to Gray image
Gray_img = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
plt.imshow(Gray_img, cmap="gray")
plt.axis(False)
plt.show()
```

```python
#Invert the image
invert_img = cv2.bitwise_not(Gray_img)
plt.imshow(Gray_img, cmap="gray")
plt.axis(False)
plt.show()
```

```python
#Blur the image
blur_img = cv2.GaussianBlur(invert_img, (111,111),0)
plt.imshow(Gray_img, cmap="gray")
plt.axis(False)
plt.show()
```

```python
#Invert the blur image
invblur_img=cv2.bitwise_not(blur_img)
```

```python
#Convert to sketch
sketch_img=cv2.divide(Gray_img,invblur_img, scale=256.0)
plt.imshow(sketch_img, cmap = "gray")
plt.axis(False)
plt.show()
```
